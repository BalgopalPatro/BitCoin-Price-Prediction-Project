### Treating the Missing data :

From the above metadata we know the our datasets conatin a lot of NaN values.

#### Visualisation of nan values

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.title('Data1',{'fontsize': 18,'fontweight' : 5})
sns.heatmap(data1.isnull(),yticklabels=False,cbar=False,cmap='Set1_r')
plt.subplot(1,2,2)
plt.title('Data2',{'fontsize': 18,'fontweight' : 5})
sns.heatmap(data2.isnull(),yticklabels=False,cbar=False,cmap='Set1_r')
```

From the above data we can see there is a lot of missing values in our datasets (in both data1 and data2)  
Now let's see the number of rows in which all the columns have `nan` values which we can delete from the datasets

```python
temp_df1 = data1.loc[(data1['Open'].isnull()) & (data1['High'].isnull())
             & (data1['Low'].isnull()) & (data1['Close'].isnull())
             & (data1['Volume_(BTC)'].isnull()) & (data1['Volume_(Currency)'].isnull())]
temp_df1.isnull().sum()
```

```python
temp_df2 = data2.loc[(data2['Open'].isnull()) & (data2['High'].isnull())
             & (data2['Low'].isnull()) & (data2['Close'].isnull())
             & (data2['Volume_(BTC)'].isnull()) & (data2['Volume_(Currency)'].isnull())]
temp_df2.isnull().sum()
```

As we can see in data1 `109069` rows and in data2 `1231878` rows have all the columns have `nan` values except Timestamp which i is equal to there respecting missing values .  
we can delete those rows by the following method :
  
`
data1.drop(temp_df1.index,inplace=True)
data2.drop(temp_df2.index,inplace=True)`  
  
Another way to treat the missing values is to fill the `nan` values with the `mean` values .

```python
for i in data1.columns:
    data1[i].fillna(data1[i].mean(),inplace=True) # Replace the nan values of each columns by the mean values in data1
for i in data2.columns:
    data2[i].fillna(data2[i].mean(),inplace=True) # Replace the nan values of each columns by the mean values in data2
```

Now all the `nan` values are replaced to their corresponding `mean` values and our data set is free of missing values .

## Modelling

### Data Preprocessing

```python
data1['Timestamp'] = data1['Timestamp'].map(dt.datetime.toordinal)
data2['Timestamp'] = data2['Timestamp'].map(dt.datetime.toordinal)
```

Now Extracting independent and dependent variables from the DataFrame .  
We have two datasets and we visualise both the Datasets and we can say that data1 is a subset of data2 so we can proceed further 
with data2 because it has more number of rows . 

```python
df = data2
```

```python
x = df['Timestamp'].values
x = x.reshape(-1,1)
x
```

```python
y = df['Weighted_Price'].values
y
```

##### Split the data into train and test 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35)
```

### Standardization of data 
Standardization means scaling the data into a form in which mean will be 0 and stadard deviation of 1

```python
print("Maximun value of x_train :",x_train.max())
print("Minimum value of x_train :",x_train.min())
```

As we can see the values are large so we will scall the data using StandScaler or MinMaxScaler

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print("After Scalling : ")
print("Maximun value of x_train :",x_train.max())
print("Minimum value of x_train :",x_train.min())
```

#### Applying LinearRegression to the DataSet

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr_acc = lr.score(x_test,y_test)
print("Accuracy (R 2 square method) :",lr_acc)
```

```python
lr_mse = mean_squared_error(y_test,lr.predict(x_test))
print("Mean Squared Error : ",lr_mse)
```

##### Visualization Regression Line

```python
plt.figure(figsize=(20,6))
plt.title("Linear Regression Line",fontsize=20)
plt.scatter(x=x_test,y=y_test,marker='*',color='lightblue',label='Actual data')
plt.plot(x_test,lr.predict(x_test),'r--',linewidth=5,label='Regression Line')
plt.legend()
plt.show()
```

###### As we can see the in some cases prediction by regression line differs a lot with the actual data because the accuracy is very low

```python
plt.figure(figsize=(20,6))
res = pd.DataFrame({'Origional' : y_test,'Pred' : lr.predict(x_test),'Error' : (y_test-lr.predict(x_test))}).head(10)
res.plot(kind='bar',figsize=(20,6))
plt.grid('-',which='major')
```

#### Applying TreeRegressor to DataSet

```python
tree_reg = DecisionTreeRegressor()
```

```python
tree_reg.fit(x_train,y_train)
y_tree_pred = tree_reg.predict(x_test)
tree_pred_score = tree_reg.score(x_test,y_test)
```

```python
print("The Accuracy for the DecisionTreeRegressor :",tree_pred_score)
```

```python
tree_mse = mean_squared_error(y_test,y_tree_pred)
print("Mean Squared Error : ",tree_mse)
```

##### Visualisation of Prediction Result 

```python
plt.figure(figsize=(20,6))
plt.title("DecisionTreeRegressor",fontsize=20)
plt.scatter(x=x_test,y=y_test,marker='*',color='lightblue',label='Actual data')
sns.lineplot(x_test.reshape(-1),y_tree_pred,linewidth=5,label='Predicted value')
plt.legend()
plt.show()
```

###### As the accuracy is more than 91% , It is clear from the above plot ,Because almost points are very close to their Actual Value

#### Applying Polynomial Regression in DataSet

```python
poly_sc = PolynomialFeatures(degree=4);
```

```python
x_train_poly = poly_sc.fit_transform(x_train)
x_test_poly = poly_sc.transform(x_test)
```

```python
poly_lr = LinearRegression()
poly_lr.fit(x_train_poly,y_train)
```

```python
y_poly_pred = poly_lr.predict(x_test_poly)
poly_lr_score = poly_lr.score(x_test_poly,y_test)
print("The Accuracy For the Polynomial Refression : ",poly_lr_score)
```

```python
poly_lr_mse = mean_squared_error(y_test,y_poly_pred)
print('Mean Squared Error :',poly_lr_mse)
```

##### Visualization of prediction Results

```python
plt.figure(figsize=(20,6))
plt.title("Polynomial Regression of Degree 2",fontsize=20)
plt.scatter(x=x_test,y=y_test,marker='*',color='lightblue',label='Actual data')
sns.lineplot(x_test.reshape(-1),y_poly_pred,linewidth=5,label='Predicted value',color='red')
plt.legend()
plt.show()
```

In the case of Polynomial Regression The Accuracy depends on the `degree` But too large value `degree` causes a overfit and too small value of `degree` causes underfit .  
Let's Do the Polynomial Regression for different values of `degree`

```python
l = []
for i in range(2,21):
    poly_f = PolynomialFeatures(degree=i);
    pr = LinearRegression()
    xp = poly_f.fit_transform(x_train)
    xtp = poly_f.fit_transform(x_test)
    pr.fit(xp,y_train)
    pr_score = pr.score(xtp,y_test)
    l.append([i,pr_score])
```

```python
score_data = pd.DataFrame(l,columns=['Degree','Accuracy Score'])
score_data
```

###### Let's try to plot the Graph with degree 30

```python
poly_f = PolynomialFeatures(degree=30);
pr = LinearRegression()
xp = poly_f.fit_transform(x_train)
xtp = poly_f.fit_transform(x_test)
pr.fit(xp,y_train)
ytp = pr.predict(xtp)
pr_score = pr.score(xtp,y_test)
print("Prediction Accuracy :",pr_score)

# Visualization of the Result

plt.figure(figsize=(20,6))
plt.title("Polynomial Regression of Degree 30",fontsize=20)
plt.scatter(x=x_test,y=y_test,marker='*',color='lightblue',label='Actual data')
sns.lineplot(x_test.reshape(-1),ytp,linewidth=5,label='Predicted value',color='red')
plt.legend()
plt.show()
```

There is another way to Predict the price of bitcoin for n next days using the regression with  the Existing bitcoin price column .  
Let's try this 

```python
n = 30 # number of future day for prediction
df = pd.DataFrame(data1.Weighted_Price)
```

```python
df['Prediction'] = df[['Weighted_Price']].shift(-n)
```

Create a independent DataSet (X)

```python
x = df['Weighted_Price'].values
```

```python
# Delete the last n rows for which we have to predict
x = x[:len(df) - n]
```

Create a Dependent DataSet (Y) 

```python
y = df['Prediction'].values
y = y[:-n] 
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

Now we will set prediction_days equal to the last 30 rows of the original data set from the price column for which we have to predict

```python
x_test = x_test.reshape(-1,1)
x_train = x_train.reshape(-1,1)
```

```python
prediction_days = df.iloc[-n:,0].values
prediction_days.shape
```

```python
lrg = LinearRegression() 
lrg.fit(x_train, y_train) 
```

```python
lrg_score = lrg.score(x_test, y_test)
y_pred = lrg.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print("Regression Accuracy: ", lrg_score)
print("Mean Squared Error: ", mse)
```

##### **As we got the very good Accuracy we can now predict the future bitcoin value of last 30 days.**

#### Visualization : 

```python
plt.figure(figsize=(20,6))
plt.title("Linear Regression ",fontsize=20)
plt.scatter(x_test.reshape(-1),y_test,marker='*',color='lightblue',label='Actual data')
sns.lineplot(x_test.reshape(-1),y_pred,linewidth=5,label='Predicted value',color='red')
plt.legend()
plt.show()
```

```python
n_predictions = lrg.predict(prediction_days.reshape(-1,1))
df.iloc[-n:,1] = n_predictions
df.iloc[-n:,:]
```
